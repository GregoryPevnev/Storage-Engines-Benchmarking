# Analysis of speed and throughput of storage engines

## Helpers and utilities

```python
import numpy as np 
import matplotlib.pyplot as plt
import linecache

storage_engines_labels = ['SQLite3', 'ForestDB', 'LevelDB']
storage_engines_labels_axis = np.arange(len(storage_engines_labels))

def analyze_speed_throughput_for(speed_throughput_info_file):
    writing_time = int(linecache.getline(speed_throughput_info_file, 4).split(' ')[3])
    reading_time = int(linecache.getline(speed_throughput_info_file, 5).split(' ')[3])

    writing_throughput = int(linecache.getline(speed_throughput_info_file, 8).split(' ')[2])
    reading_throughput = int(linecache.getline(speed_throughput_info_file, 9).split(' ')[2])

    return writing_time, reading_time, writing_throughput, reading_throughput

sqlite3_performance_metrics = analyze_speed_throughput_for('../benchmarks/sqlite3/speed_throughput')
forestdb_performance_metrics = analyze_speed_throughput_for('../benchmarks/forestdb/speed_throughput')
leveldb_performance_metrics = analyze_speed_throughput_for('../benchmarks/leveldb/speed_throughput')
```

## Speed comparison chart

```python
write_times = [sqlite3_performance_metrics[0], forestdb_performance_metrics[0], leveldb_performance_metrics[0]]
read_times = [sqlite3_performance_metrics[1], forestdb_performance_metrics[1], leveldb_performance_metrics[1]]

plt.figure(figsize = (20, 10))

plt.bar(storage_engines_labels_axis - 0.2, read_times, 0.4, color='blue', label = 'total time reading in ms')
plt.bar(storage_engines_labels_axis + 0.2, write_times, 0.4, color='red', label = 'total time writing in ms')

plt.xticks(storage_engines_labels_axis, storage_engines_labels)

plt.title('Reading / Writing Speed')
plt.xlabel('Storage Engines')
plt.ylabel('Total time reading / writing in ms')
plt.legend()
plt.show()

print('SQLite')
print('Writing all the data in ' + str(sqlite3_performance_metrics[0]) + 'ms and reading it in ' + str(sqlite3_performance_metrics[1]) + 'ms\n')

print('ForestDB')
print('Writing all the data in ' + str(forestdb_performance_metrics[0]) + 'ms and reading it in ' + str(forestdb_performance_metrics[1]) + 'ms\n')

print('LevelDB')
print('Writing all the data in ' + str(leveldb_performance_metrics[0]) + 'ms and reading it in ' + str(leveldb_performance_metrics[1]) + 'ms\n')
```

## Throughput comparison chart

```python
write_times = [sqlite3_performance_metrics[2], forestdb_performance_metrics[2], leveldb_performance_metrics[2]]
read_times = [sqlite3_performance_metrics[3], forestdb_performance_metrics[3], leveldb_performance_metrics[3]]

plt.figure(figsize = (20, 10))

plt.bar(storage_engines_labels_axis - 0.2, read_times, 0.4, color='blue', label = 'throughput of reading in bytes/ms')
plt.bar(storage_engines_labels_axis + 0.2, write_times, 0.4, color='red', label = 'throughput of writing in bytes/ms')

plt.xticks(storage_engines_labels_axis, storage_engines_labels)

plt.title('Reading / Writing Throughput')
plt.xlabel('Storage Engines')
plt.ylabel('Throughput of reading / writing in bytes/ms')
plt.legend()
plt.show()

print('SQLite')
print('Writing throuhgput of ' + str(sqlite3_performance_metrics[2]) + 'bytes/ms and reading throughput of ' + str(sqlite3_performance_metrics[3]) + 'bytes/ms\n')

print('ForestDB')
print('Writing throuhgput of ' + str(sqlite3_performance_metrics[2]) + 'bytes/ms and reading throughput of ' + str(sqlite3_performance_metrics[3]) + 'bytes/ms\n')

print('LevelDB')
print('Writing throuhgput of ' + str(sqlite3_performance_metrics[2]) + 'bytes/ms and reading throughput of ' + str(sqlite3_performance_metrics[3]) + 'bytes/ms\n')
```
